## Matplotlib
Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python. <br>
https://matplotlib.org/

```python
import matplotlib.pyplot as plt


%matplotlib inline  
#magic command. See https://ipython.readthedocs.io/en/stable/interactive/magics.html
```

```python
#Generate random data
time = np.linspace(0, 120, 600)
data = np.sin(np.arange(0, 30, 0.05)) + np.random.randn(600)/2
```

```python
#Create a figure with one axis
fig, ax = plt.subplots(figsize = (10,4))

#Plot the data
ax.plot(time, data, '-', c = 'blue', lw = 1, label ='sample 1')
ax.plot(time, data - 6 - np.random.randn(600), '.-', c = 'red', lw = 1, label ='sample 2')


#Label axis
ax.set_xlabel('Time (sec)', fontsize = 15)
ax.set_xticks(np.arange(0,130,30))

#Show legend
plt.legend(loc = 1)

#decide the range
plt.xlim(0,120)
plt.title('Test', fontsize = 20)

plt.text(10,-2,"sample 1", color='blue')

#If you want to save the image...
#plt.savefig(r'Sample.png', dpi = 300)
```

```python
#Slightly different style.
fig, ax = plt.subplots(figsize = (10,4))
ax.plot(time, data, '.', c = 'b', label ='sample 1', alpha = 0.4)
ax.plot(time, np.sin(np.arange(0, 30, 0.05)), lw = 2, c = 'b')

ax.set_xlabel('Time', fontsize = 15)
ax.set_xticks(np.arange(0,130,30))

plt.legend()

plt.xlim(0, 120)

plt.title('Test', fontsize = 20)

#plt.savefig('Sample.png', dpi = 300)
```

```python
#xkcd!! https://xkcd.com/
with plt.xkcd():
    fig, ax = plt.subplots(figsize = (10,4))
    ax.plot(time, data, '.', c = 'b', label ='sample 1', alpha = 0.4)
    ax.plot(time, np.sin(np.arange(0, 30, 0.05)), lw = 2, c = 'b')

    ax.set_xlabel('Time', fontsize = 15)
    ax.set_xticks(np.arange(0,130,30))

    plt.legend()

    plt.xlim(0, 120)

    plt.title('Test', fontsize = 20)

#plt.savefig('Sample.png', dpi = 300)
```

## Seaborn

Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.

```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
#load sample data and show how it looks
penguins_df = sns.load_dataset('penguins')
penguins_df
```

```python
#distribution plot
sns.displot(penguins_df['flipper_length_mm'], kde = 0, bins = 20)
```

```python
sns.displot(penguins_df, x="flipper_length_mm", hue="species")
```

```python
sns.jointplot(x = 'flipper_length_mm', y = 'body_mass_g', data = penguins_df, kind ='reg')
```

```python
sns.jointplot(x = 'flipper_length_mm', y = 'body_mass_g', data = penguins_df, hue = 'species')
```

```python
sns.boxplot(x = 'species', y = 'body_mass_g', data = penguins_df, hue ='sex')
```

```python
sns.swarmplot(x = 'species', y = 'body_mass_g', data = penguins_df, hue ='sex')
```
